```python
import pandas as pd
```

```python
df = pd.read_csv("../data/processed/dengue_departamental_semanal_completo.csv")
```

```python
df.head()
```

```python
df["fecha"] = pd.to_datetime(
    df["ano"].astype(str) + "-W" +
    df["semana"].astype(str) + "-1",
    format="%Y-W%W-%w"
)
```

```python
df[["ano", "semana", "fecha"]].head(10)
```

```python
df = df.sort_values(["departamento", "fecha"])
```

```python
df.to_csv(
    "../data/processed/dengue_departamental_semanal_con_fecha.csv",
    index=False
)
```
