```python
import pandas as pd
```

```python
df = pd.read_csv("../data/raw/dengue_minsa.csv", encoding="latin-1")
df.head()
```

```python
df.columns
```

```python
df = df[df["enfermedad"].str.upper() == "DENGUE"]
```

```python
departamentos = [
    "LORETO",
    "PIURA",
    "SAN MARTIN",
    "LA LIBERTAD",
    "LIMA"
]

df["departamento"] = df["departamento"].str.upper()
df = df[df["departamento"].isin(departamentos)]
```

```python
df = df[(df["ano"] >= 2018) & (df["ano"] <= 2023)]
```

```python
df_agg = (
    df
    .groupby(["departamento", "ano", "semana"])
    .size()
    .reset_index(name="casos_dengue")
)
```

```python
df_agg.head()
```

```python
df_agg.to_csv("../data/processed/dengue_departamental_semanal.csv", index=False)
```

```python
departamentos = df_agg["departamento"].unique()
anios = df_agg["ano"].unique()
semanas = range(1, 53)  # semanas epidemiolÃ³gicas
```

```python
import itertools

estructura_completa = pd.DataFrame(
    list(itertools.product(departamentos, anios, semanas)),
    columns=["departamento", "ano", "semana"]
)
```

```python
df_completo = estructura_completa.merge(
    df_agg,
    on=["departamento", "ano", "semana"],
    how="left"
)
```

```python
df_completo["casos_dengue"] = df_completo["casos_dengue"].fillna(0).astype(int)
```

```python
df_completo[
    (df_completo["departamento"] == "LA LIBERTAD") &
    (df_completo["ano"] == 2018)
].head(15)
```

```python
df_completo.to_csv(
    "../data/processed/dengue_departamental_semanal_completo.csv",
    index=False
)
```
