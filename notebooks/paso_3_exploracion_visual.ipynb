```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/processed/dengue_departamental_semanal_con_fecha.csv")
```

```python
df["fecha"] = pd.to_datetime(df["fecha"])
```

```python
dep = "LA LIBERTAD"
df_dep = df[df["departamento"] == dep]
```

```python
plt.figure(figsize=(12, 5))
plt.plot(df_dep["fecha"], df_dep["casos_dengue"])
plt.title(f"Casos de dengue por semana - {dep}")
plt.xlabel("Fecha")
plt.ylabel("Casos de dengue")
plt.grid(True)
plt.show()
```

```python
deps = ["LA LIBERTAD", "PIURA"]

plt.figure(figsize=(12, 5))
for d in deps:
    temp = df[df["departamento"] == d]
    plt.plot(temp["fecha"], temp["casos_dengue"], label=d)

plt.title("Comparaci√≥n de casos de dengue por semana")
plt.xlabel("Fecha")
plt.ylabel("Casos")
plt.legend()
plt.grid(True)
plt.show()
```
